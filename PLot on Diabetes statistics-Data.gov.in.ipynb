### Line PLot

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)

# fake up some data
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))
```

```python
data.boxplot(by='Categories',column=['NFHS III(2005-06)-Number of women in age group 15-49 per 100-000 reported Diabetes-Total'],grid=False)
```

### Bar Plot

```python
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(data['Categories'], data['NFHS III(2005-06)-Number of women in age group 15-49 per 100-000 reported Diabetes-Total'], 
             color ='maroon',width = 0.4)
 
plt.xlabel("Categories")
plt.ylabel("NFHS III(2005-06)-Number of women in age group 15-49 per 100-000 reported Diabetes-Total")
plt.title("Diabetes Statistics Indian Women")
plt.show()
```
